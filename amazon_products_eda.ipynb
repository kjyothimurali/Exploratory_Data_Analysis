```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
path="/content/drive/MyDrive/amazon_products_dataset.csv"
df=pd.read_csv(path)
```

```python
print(df.head())
print(df.tail())
```

```python
print(df.shape)
print(df.info)
```

```python
print(df['category'].unique())
df.isnull().sum()
```

```python
df['discount_price'].fillna(df['discount_price'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['rating'].fillna(df['rating'].mean(),inplace=True)
df['actual_price'].fillna(df['actual_price'].mean(),inplace=True)
df['category'].fillna('unknown',inplace=True)
df['rating_count'].fillna(df['rating_count'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.dtypes
```

```python
df=df.reset_index(drop=True)
```

```python
df.groupby('category')['actual_price'].nlargest(5)
```

```python
df.groupby('brand').size().sort_values(ascending=False).head(1)
```

```python
df.groupby('brand')['discount_price'].mean()
```

```python
filtered = df[(df['rating'] >= 4.5) & (df['rating_count'] >= 2000)].reset_index(drop=True)
filtered.head()
```

```python
df.groupby('category', observed=True)['actual_price'].mean().sort_values(ascending=False)
```

```python
import numpy as np
df['discount_percent'] = ((df['actual_price'] - df['discount_price']) / df['actual_price']) * 100
df['price_category'] = df['actual_price'].apply(
    lambda x: 'Budget' if x < 500
    else ('Midrange' if x < 2000
          else 'Premium')
)

df['popularity_score'] = df['rating'] * np.log(df['rating_count'] + 1)
print(df.head())
```

```python
df.groupby('category')['popularity_score'].nlargest(10)
```

```python
filter_df=df[df['price_category'].apply(lambda x:x=='Premium') & df['discount_percent'].apply(lambda y:y>40)]
filter_df['category']
```

```python
bluetooth_df = df[df['title'].str.contains('Bluetooth', case=False, na=False)]
bluetooth_df.head()
```
